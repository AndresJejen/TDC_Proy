```python
y_max = 34
y_min = 1.1
print(f"Average Max Y: {y_max}, Average Min Y: {y_min}")
```

```python
import math 

ud = 2
y2 = 30.4
y1 = 7.2
t2 = 103.35
t1 = 102.2

Tu = 115.075 - 100.575

ypp = y_max - y_min
e_h = 3
a = ypp/2

ku = 4*ud/(math.pi*math.sqrt(a**2 - e_h**2))
print("ku:", ku)
w_u = 2*math.pi/Tu
print("w_u:", w_u)

b = (y2 - y1) / ((t2 - t1) * ud)
print("b:", b)
tau = math.sqrt( ( (ku*b)**2 - (w_u)**2 ) ) / (w_u)**2
print("tau:", tau)
L = ((math.pi/2) - math.atan(tau*w_u)) / w_u
print("L:", L)

print("*"*50)
print("ZIEGLER-NICHOLS")
kp = 0.6*ku
print("kp:", kp)
ki = kp/(0.5*Tu)
print("ki:", ki) 
kd = 0.125*Tu*kp
print("kd:", kd)

print("*"*50)
print("AMSTRONG")
kp = (0.37/(b*L)) + 0.02*(tau/(b*L**2))
print("kp:", kp)
ki = (0.03/(b*L)) + 0.0012*(tau/(b*L**3))
print("ki:", ki)
kd = (0.16/b) + 0.28*(tau/(b*L))
print("kd:", kd)
```
